```python
with open('белеет_парус.txt', encoding='utf-8') as f:
    data = f.read()
```

```python
import re
```

```python
lst = []
lst = re.findall(r'[а-яА-Я]+',data)
lst
```

```python
dct_lo = {k:lst.count(k) for k in lst if k[0] == k[0].lower()}
dct_up = {k.lower():lst.count(k) for k in lst if k[0] == k[0].upper()}
```

```python
dct = {}
dct.update(dct_lo)
for k in dct_up:
    dct[k] = dct.get(k,0) + dct_up[k]
dct
```
